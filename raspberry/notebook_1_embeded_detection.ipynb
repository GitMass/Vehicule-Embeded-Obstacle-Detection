```python
# Path to your manually downloaded dataset
source_path = r"datasets\bdd100k_obj"

# verify path
if not os.path.exists(source_path):
    print(f"Error: The path '{source_path}' does not exist on this computer.")
    print(f"Current working directory is: {os.getcwd()}")
else:
    print(f"Path found! Checking contents...")
    if "images" in os.listdir(source_path) and "labels" in os.listdir(source_path):
        print("Structure looks correct (found 'images' and 'labels').")
    else:
        print(f"Error: The folder exists, but I don't see 'images' and 'labels' inside it.")
        print(f"Found files: {os.listdir(source_path)}")

# Load dataset
dataset = foz.load_zoo_dataset(
    "bdd100k",
    split="validation",
    max_samples=1000,
    # This tells FiftyOne where the raw images/labels are
    source_dir=source_path
)

print(f"Dataset loaded successfully : Lenght {len(dataset)}")
```
