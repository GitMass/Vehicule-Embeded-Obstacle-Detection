# Download a dataset subset

```python
import fiftyone as fo
import fiftyone.zoo as foz

# Download a tiny subset of COCO-2017 (only cars and people)
# This is perfect for testing your "Stage 2" simulation logic
dataset = foz.load_zoo_dataset(
    "coco-2017",
    split="validation",       # Use 'validation' for a small, quick download
    label_types=["detections"],
    classes=["person", "car", "bicycle", "truck", "bus"], # Your "Obstacles"
    max_samples=500,          # Limit to 500 images for speed
)

# Export to a folder structure your YOLO/EfficientDet model understands
export_dir = "./datasets"
dataset.export(
    export_dir=export_dir,
    dataset_type=fo.types.YOLOv5Dataset, # Standard format (Images + Txt labels)
)

print(f"Dataset ready at {export_dir}")
```
